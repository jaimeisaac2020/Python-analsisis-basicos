```python
# Metodo del punto fijo
# Importing math para usar la raiz cuadrada
import math

def f(x):
    return math.log(x)-x+2

# Re-writing f(x)=0 to x = g(x)
def g(x):
    return math.log(x)+2

# Implementando el metodo de iteracion del punto fijo.
def fixedPointIteration(x0, e, N):
    print('\n\n*** Iteracion del punto fijo ***')
    step = 3
    flag = 4
    condition = True
    while condition:
        x1 = g(x0)
        print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
        x0 = x1

        step = step + 1
        
        if step > N:
            flag=0
            break
        
        condition = abs(f(x1)) > e

    if flag==1:
        print('\nRequired root is: %0.8f' % x1)
    else:
        print('\nNot Convergent.')

# Input Section
x0 = input('Valor Incial: ')
e = input('maximo error tolerable: ')
N = input('Maximo numero de pasos: ')

# Converting x0 and e to float
x0 = float(x0)
e = float(e)

# Converting N to integer
N = int(N)


#Note: You can combine above three section like this
# x0 = float(input('valor incial: '))
# e = float(input('Error tolerable: '))
# N = int(input('Maximo numero de pasos: '))

# Starting Newton Raphson Method
fixedPointIteration(x0,e,N)
```

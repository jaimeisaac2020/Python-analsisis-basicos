```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Generar valores entre -3 y 3 en intervalos de 0.1
T = np.arange(-3, 3.1, 0.1)

# Función de densidad para una t de Student con gl=10
f1t = t.pdf(T, df=10)

# Función de densidad para una t de Student con gl=40
f2t = t.pdf(T, df=40)

# Función de densidad para una t de Student con gl=60
f3t = t.pdf(T, df=60)

# Crear el gráfico
plt.plot(T, f1t, label='df=10', linewidth=1.5)
plt.plot(T, f2t, label='df=40', linewidth=1.5, linestyle='dashed', color='red')
plt.plot(T, f3t, label='df=60', linewidth=1.5, linestyle='dashed', color='blue')

# Configuraciones adicionales
plt.ylim(0, 0.40)
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
from scipy.stats import t

# Parámetros: X=2, grados de libertad=10
x = 2
df = 10

# Función de distribución para X=2 en una t de Student con gl=10
cdf_value = t.cdf(x, df)
print("P(t10 ≤ 2) =", cdf_value)

# 1 - F(x) para 2 en una t de Student con gl=10
complement_cdf_value = 1 - cdf_value
print("P(t10 ≥ 2) =", complement_cdf_value)
```

```python
from scipy.stats import t

# Parámetros: Percentil=0.25, grados de libertad=10
percentil = 0.25
df = 10

# Puntuación t de Student con gl=10 para F(x) = 0.25
quantile_value = t.ppf(percentil, df)
print(f"Puntuación t de Student con gl=10 para F(x) = {percentil}: {quantile_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generar valores entre 0 y 5 en intervalos de 0.1
X = np.arange(0, 5.1, 0.1)

# Función de densidad para una F de Snedecor con gl 5 y 15
f1f = f.pdf(X, 5, 15)

# Función de densidad para una F de Snedecor con gl 10 y 20
f2f = f.pdf(X, 10, 20)

# Función de densidad para una F de Snedecor con gl 20 y 30
f3f = f.pdf(X, 20, 30)

# Crear el gráfico
plt.plot(X, f1f, label='F(5, 15)', linewidth=1.5)
plt.plot(X, f2f, label='F(10, 20)', linewidth=1.5, linestyle='dashed', color='red')
plt.plot(X, f3f, label='F(20, 30)', linewidth=1.5, linestyle='dashed', color='blue')

# Configuraciones adicionales
plt.ylim(0, 1.05)
plt.xlabel('Valores')
plt.ylabel('Densidad')

# Añadir etiquetas al gráfico
plt.text(2.9, 0.17, 'F(5, 15)', color='black')
plt.text(2.3, 0.34, 'F(10, 20)', color='red')
plt.text(1.8, 0.87, 'F(20, 30)', color='blue')

# Mostrar el gráfico
plt.legend()
plt.show()
```

A continuación se muestran ejemplos sobre cómo asignar áreas de probabilidad a ciertos valores de F de Snedecor con 10 y 15 grados de libertad. Se solicita a $R$ la función de distribución $F\left(x_i\right)$ para una $X_i=2,544$, así como la probabilidad asociada al lado derecho de esa misma puntuación $\left(F_{10,15}=2,544\right)$ :

```python
from scipy.stats import f

# Parámetros: Valor=2.544, grados de libertad=10 y 15
valor = 2.544
df1 = 10
df2 = 15

# Función de distribución acumulativa para F con gl 10 y 15
cdf_value = f.cdf(valor, df1, df2)
print(f"P(F(10, 15) ≤ 2.544) = {cdf_value}")
```

```python
from scipy.stats import f

# Parámetros: Valor=2.544, grados de libertad=10 y 15
valor = 2.544
df1 = 10
df2 = 15

# 1 - F(x) para F con gl 10 y 15
complement_cdf_value = f.sf(valor, df1, df2)
print(f"P(F(10, 15) ≥ 2.544) = {complement_cdf_value}")
```

Como puede verse: $P\left(F_{10,15} \leq 2,544\right)=0,95$ y $P\left(F_{10,15} \geq 2,544\right)=0,05$.
También es posible obtener la puntuación asociada a cierta función de distribución. Por ejemplo, continuando con el ejemplo de una $\mathrm{F}$ de Snedecor con 10 y 15 grados de libertad, el $C_{25}$ corresponde a la puntuación $F_{10,15}=0,65$ :

```python
from scipy.stats import f

# Parámetros: Percentil=0.25, grados de libertad=10 y 15
percentil = 0.25
df1 = 10
df2 = 15

# Puntuación F con gl 10 y 15 para F(x) = 0.25
quantile_value = f.ppf(percentil, df1, df2)
print(f"Puntuación F con gl 10 y 15 para F(x) = {percentil}: {quantile_value}")
```

Veremos a continuación un ejemplo más sofisticado. Simularemos 100 matrices de datos de tamaño nueve para una distribución $\mathrm{N}(100 ; 15)$.
```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de una distribución normal con media 100 y desviación estándar 15
muestra = np.random.normal(100, 15, 900)

# Redimensionar la muestra a una matriz de 9 filas y 100 columnas
muestra = np.reshape(muestra, (9, 100))

# Calcular las medias por columna
medias = np.mean(muestra, axis=0)

# Crear el histograma de medias
plt.hist(medias, bins=20, edgecolor='black')
plt.xlabel('Medias')
plt.ylabel('Frecuencia')
plt.title('Histograma de Medias')

# Añadir leyenda
plt.legend([f'Media = {np.mean(medias):.2f}\nSd = {np.std(medias):.2f}'], loc='upper right')

# Mostrar el gráfico
plt.show()
```

Ejercicio 1: Binomial
El $70 \%$ de pacientes que sufren una determinada enfermedad se curan al aplicarles un tratamiento. Si tomamos una m.a.s. de 10 pacientes con esa enfermedad y pasan el tratamiento...
1. ¿Cuál es la probabilidad de que se recuperen todos?
2. ¿Cuál es la probabilidad de que se recuperen al menos la mitad?
3. ¿Cuál es la probabilidad de que se recuperen como máximo 6 pacientes?
4. ¿Cuál es la probabilidad de que se recuperen entre 2 y 4 (ambos inclusive)?

```python
from scipy.stats import binom

# Parámetros
n = 10  # tamaño de la muestra
p = 0.7  # probabilidad de recuperación

# 1. Probabilidad de que se recuperen todos
prob_todos = binom.pmf(10, n, p)
print(f"1. Probabilidad de recuperación de todos: {prob_todos:.4f}")

# 2. Probabilidad de que se recuperen al menos la mitad
prob_al_menos_mitad = 1 - binom.cdf(4, n, p)
print(f"2. Probabilidad de recuperación al menos la mitad: {prob_al_menos_mitad:.4f}")

# 3. Probabilidad de que se recuperen como máximo 6 pacientes
prob_maximo_6 = binom.cdf(6, n, p)
print(f"3. Probabilidad de recuperación como máximo 6 pacientes: {prob_maximo_6:.4f}")

# 4. Probabilidad de que se recuperen entre 2 y 4 (ambos inclusive)
prob_entre_2_y_4 = binom.cdf(4, n, p) - binom.cdf(1, n, p)
print(f"4. Probabilidad de recuperación entre 2 y 4 pacientes: {prob_entre_2_y_4:.4f}")
```

# Ejercicio 2: Normal
La variable CI se distribuye $\mathrm{N}(100 ; 15)$ en la población de estudiantes universitarios. Se selecciona una m.a.s. de 200 sujetos. Si se extrae un sujeto al azar. .
1. ¿Cuál es la probabilidad de que puntúe en CI como máximo 120 ?
2. ¿Cuál es la probabilidad de que puntúe en CI al menos 95?
3. ¿Qué puntuación en CI tiene una probabilidad acumulada de 0,5793?
4. ¿Qué puntuación en CI ocupa el centil 65?

```python
from scipy.stats import norm

# Parámetros
media = 100
desviacion_estandar = 15
tamano_muestra = 200

# 1. Probabilidad de que puntúe en CI como máximo 120
prob_maximo_120 = norm.cdf(120, media, desviacion_estandar)
print(f"1. Probabilidad de puntuación en CI como máximo 120: {prob_maximo_120:.4f}")

# 2. Probabilidad de que puntúe en CI al menos 95
prob_al_menos_95 = 1 - norm.cdf(95, media, desviacion_estandar)
print(f"2. Probabilidad de puntuación en CI al menos 95: {prob_al_menos_95:.4f}")

# 3. Puntuación en CI con probabilidad acumulada de 0.5793
puntuacion_acumulada_05793 = norm.ppf(0.5793, media, desviacion_estandar)
print(f"3. Puntuación en CI con probabilidad acumulada de 0.5793: {puntuacion_acumulada_05793:.4f}")

# 4. Puntuación en CI que ocupa el centil 65
puntuacion_centil_65 = norm.ppf(0.65, media, desviacion_estandar)
print(f"4. Puntuación en CI que ocupa el centil 65: {puntuacion_centil_65:.4f}")
```

#Ejercicio 3: Chi-cuadrado de Pearson
La variable $X$ se distribuye según el modelo chi-cuadrado con 10 grados de libertad. Si se extrae un sujeto al azar...
1. Calcule la probabilidad de que no supere el valor 9,342
2. Calcule la puntuación $X$ que corresponde a la probabilidad acumulada de 0,70
3. Calcule la probabilidad de que la variable $\mathrm{X}$ adopte como mínimo el valor 15,987

```python
from scipy.stats import chi2

# Grados de libertad
grados_libertad = 10

# 1. Probabilidad de que no supere el valor 9.342
prob_no_supere_9342 = chi2.cdf(9.342, df=grados_libertad)
print(f"1. Probabilidad de que no supere el valor 9.342: {prob_no_supere_9342:.4f}")

# 2. Puntuación X que corresponde a la probabilidad acumulada de 0.70
puntuacion_acumulada_070 = chi2.ppf(0.70, df=grados_libertad)
print(f"2. Puntuación X que corresponde a la probabilidad acumulada de 0.70: {puntuacion_acumulada_070:.4f}")

# 3. Probabilidad de que la variable X adopte como mínimo el valor 15.987
prob_minimo_15987 = 1 - chi2.cdf(15.987, df=grados_libertad)
print(f"3. Probabilidad de que la variable X adopte como mínimo el valor 15.987: {prob_minimo_15987:.4f}")
```

1. Calcule la probabilidad de que no supere el valor 2,06
2. Calcule el valor de $X$ tal que la probabilidad de obtener como máximo ese valor sea 0,70

```python
from scipy.stats import t

# Grados de libertad
grados_libertad = 25

# 1. Probabilidad de que no supere el valor 2.06
prob_no_supere_206 = t.cdf(2.06, df=grados_libertad)
print(f"1. Probabilidad de que no supere el valor 2.06: {prob_no_supere_206:.4f}")

# 2. Valor de X tal que la probabilidad de obtener como máximo ese valor sea 0.70
valor_maximo_070 = t.ppf(0.70, df=grados_libertad)
print(f"2. Valor de X tal que la probabilidad de obtener como máximo ese valor sea 0.70: {valor_maximo_070:.4f}")

#3. Si dividimos la distribución en cuatro partes iguales, ¿Qué valores generan dicha partición?
from scipy.stats import t

# Grados de libertad
grados_libertad = 25

# Encontrar los cuartiles
cuartil_1 = t.ppf(0.25, df=grados_libertad)
cuartil_2 = t.ppf(0.50, df=grados_libertad)  # Mediana
cuartil_3 = t.ppf(0.75, df=grados_libertad)

print(f"Cuartil 1: {cuartil_1:.4f}")
print(f"Cuartil 2 (Mediana): {cuartil_2:.4f}")
print(f"Cuartil 3: {cuartil_3:.4f}")
```

#Ejercicio 5: F de Snedecor
La variable $X$ se distribuye según el modelo $F_{7,8}$ Si se extrae un sujeto al azar...
1. Calcule la probabilidad de que no supere el valor 3,5
2. Calcule el valor $X$ tal que la probabilidad de obtener como mínimo ese valor sea 0,75
3. Calcule el valor de $X$ tal que la probabilidad de obtener como máximo ese valor sea 0,90
4. ¿Qué valor corresponde al primer cuartil?

```python
from scipy.stats import f

# Grados de libertad
df1 = 7
df2 = 8

# 1. Probabilidad de que no supere el valor 3.5
prob_no_supere_35 = f.cdf(3.5, df1, df2)
print(f"1. Probabilidad de que no supere el valor 3.5: {prob_no_supere_35:.4f}")

# 2. Valor X tal que la probabilidad de obtener como mínimo ese valor sea 0.75
valor_minimo_075 = f.ppf(0.75, df1, df2)
print(f"2. Valor X tal que la probabilidad de obtener como mínimo ese valor sea 0.75: {valor_minimo_075:.4f}")

# 3. Valor de X tal que la probabilidad de obtener como máximo ese valor sea 0.90
valor_maximo_090 = f.ppf(0.90, df1, df2)
print(f"3. Valor X tal que la probabilidad de obtener como máximo ese valor sea 0.90: {valor_maximo_090:.4f}")

# 4. Valor correspondiente al primer cuartil
cuartil_1 = f.ppf(0.25, df1, df2)
print(f"4. Valor correspondiente al primer cuartil: {cuartil_1:.4f}")
```

| :---: | :---: | :---: | :---: |
| N(0;1) |  |  |  |
| t_(1) |  |  |  |
| t_(10) |  |  |  |

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Parámetros
tamano_muestra = 100
num_muestras = 50

# Funciones de densidad
x = np.linspace(-5, 5, 1000)
pdf_normal = norm.pdf(x, 0, 1)
pdf_t1 = t.pdf(x, 1)
pdf_t10 = t.pdf(x, 10)

# Gráfica de las funciones de densidad
plt.plot(x, pdf_normal, label='N(0, 1)')
plt.plot(x, pdf_t1, label='t(1)')
plt.plot(x, pdf_t10, label='t(10)')
plt.title('Comparación de Distribuciones')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()

# Centil 95
centil_95_normal = norm.ppf(0.95, 0, 1)
centil_95_t1 = t.ppf(0.95, 1)
centil_95_t10 = t.ppf(0.95, 10)

# Muestras y estadísticas
muestras_normal = np.random.normal(0, 1, size=(num_muestras, tamano_muestra))
muestras_t1 = np.random.standard_t(1, size=(num_muestras, tamano_muestra))
muestras_t10 = np.random.standard_t(10, size=(num_muestras, tamano_muestra))

media_muestral_normal = np.mean(muestras_normal, axis=1)
media_muestral_t1 = np.mean(muestras_t1, axis=1)
media_muestral_t10 = np.mean(muestras_t10, axis=1)

desviacion_muestral_normal = np.std(muestras_normal, axis=1, ddof=1)
desviacion_muestral_t1 = np.std(muestras_t1, axis=1, ddof=1)
desviacion_muestral_t10 = np.std(muestras_t10, axis=1, ddof=1)

# Tabla de resultados
tabla_resultados = [
    ['N(0, 1)', centil_95_normal, np.mean(media_muestral_normal), np.std(media_muestral_normal, ddof=1)],
    ['t(1)', centil_95_t1, np.mean(media_muestral_t1), np.std(media_muestral_t1, ddof=1)],
    ['t(10)', centil_95_t10, np.mean(media_muestral_t10), np.std(media_muestral_t10, ddof=1)]
]

# Mostrar resultados
print("\nTabla de Resultados:")
print("Distribución\tCentil 95\tMedia Muestral\tDesviación Muestral")
for resultado in tabla_resultados:
    print(f"{resultado[0]}\t{resultado[1]:.4f}\t{resultado[2]:.4f}\t{resultado[3]:.4f}")
```

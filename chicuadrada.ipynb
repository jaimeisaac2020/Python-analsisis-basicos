````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Generar datos aleatorios distribuidos según la chi cuadrada
df = 5  # Grados de libertad
data = np.random.chisquare(df, 1000)

# Crear un histograma para visualizar la distribución
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')

# Calcular la función de densidad de probabilidad (PDF) teórica
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = chi2.pdf(x, df)

# Graficar la PDF teórica
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: df = %.2f" % df
plt.title(title)

plt.show()

# Realizar un test de chi cuadrada
expected_values = chi2.pdf(data, df)
chi_squared_stat = np.sum(((data - expected_values) ** 2) / expected_values)
critical_value = chi2.ppf(0.95, df)  # Obtener el valor crítico para un nivel de significancia del 95%

# Imprimir resultados del test
print("Chi-squared Statistic:", chi_squared_stat)
print("Critical Value:", critical_value)
print("p-value:", 1 - chi2.cdf(chi_squared_stat, df))
````

### Ejercicio 1: Intervalo de confianza para la varianza de una población.

```python
import numpy as np
from scipy.stats import chi2

# Datos de muestra
datos = np.array([14, 15, 16, 12, 10, 11, 13, 14, 15, 13])

nivel_confianza = 0.95

# Calcular la media y la varianza de la muestra
media_muestral = np.mean(datos)
varianza_muestral = np.var(datos, ddof=1)

# Calcular el intervalo de confianza para la varianza
alfa = 1 - nivel_confianza
chi2_inf = chi2.ppf(alfa / 2, grados_libertad)
chi2_sup = chi2.ppf(1 - alfa / 2, grados_libertad)

intervalo_confianza = [
    (grados_libertad * varianza_muestral) / chi2_sup,
    (grados_libertad * varianza_muestral) / chi2_inf
]

print(f"Intervalo de confianza para la varianza: {intervalo_confianza}")
```

Ejemplo: Prueba de bondad de ajuste para una distribución discreta.

```python
import numpy as np
from scipy.stats import chi2

# Datos observados
observados = np.array([18, 22, 16, 5, 9])

# Frecuencias esperadas bajo la hipótesis nula
esperados = np.array([15, 20, 15, 10, 10])

# Grados de libertad (número de categorías - 1)
grados_libertad = len(observados) - 1

# Calcular el estadístico de prueba (chi cuadrada)
chi_squared_stat = np.sum((observados - esperados)**2 / esperados)

# Calcular el valor crítico para el nivel de significancia deseado
nivel_significancia = 0.05
valor_critico = chi2.ppf(1 - nivel_significancia, grados_libertad)

# Imprimir el resultado de la prueba
print(f"Estadístico de prueba (chi cuadrada): {chi_squared_stat}")
print(f"Valor crítico para un nivel de significancia de {nivel_significancia}: {valor_critico}")

# Realizar la prueba de hipótesis
p_valor = 1 - chi2.cdf(chi_squared_stat, grados_libertad)
print(f"P-valor: {p_valor}")

# Decidir si rechazar o no la hipótesis nula
if chi_squared_stat > valor_critico:
    print("Se rechaza la hipótesis nula. Hay evidencia para concluir que los datos no siguen la distribución esperada.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. Los datos podrían seguir la distribución esperada.")
```

En este ejemplo, estamos realizando una prueba de bondad de ajuste para verificar si los datos observados siguen una distribución específica (dada por las frecuencias esperadas). Calculamos el estadístico de prueba chi cuadrada y comparamos con el valor crítico para tomar una decisión sobre la hipótesis nula.

Ajusta los datos y las frecuencias esperadas según tus necesidades y asegúrate de entender cada paso del código.

````python
import numpy as np
from scipy.stats import chi2

# Datos observados
observados = np.array([18, 22, 16, 5, 9])

# Frecuencias esperadas bajo la hipótesis nula
esperados = np.array([15, 20, 15, 10, 10])

# Grados de libertad (número de categorías - 1)
grados_libertad = len(observados) - 1

# Calcular el estadístico de prueba (chi cuadrada)
chi_squared_stat = np.sum((observados - esperados)**2 / esperados)

# Calcular el valor crítico para el nivel de significancia deseado
nivel_significancia = 0.05
valor_critico = chi2.ppf(1 - nivel_significancia, grados_libertad)

# Imprimir el resultado de la prueba
print(f"Estadístico de prueba (chi cuadrada): {chi_squared_stat}")
print(f"Valor crítico para un nivel de significancia de {nivel_significancia}: {valor_critico}")

# Realizar la prueba de hipótesis
p_valor = 1 - chi2.cdf(chi_squared_stat, grados_libertad)
print(f"P-valor: {p_valor}")

# Decidir si rechazar o no la hipótesis nula
if chi_squared_stat > valor_critico:
    print("Se rechaza la hipótesis nula. Hay evidencia para concluir que los datos no siguen la distribución esperada.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. Los datos podrían seguir la distribución esperada.")
````

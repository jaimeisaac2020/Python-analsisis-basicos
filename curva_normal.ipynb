```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu=998.8
sigma=73.1
x1=990
x2=1100
```

```python
z1=(x1-mu)/sigma
z2=(x2-mu)/sigma
print(z1,z2)
```

```python
x=np.arange(z1,z2,0.0001)
x2=np.arange(-10,10,0.0001)
y=norm.pdf(x,0,1)
y2=norm.pdf(x2,0,1)
fig,ax=plt.subplots(figsize=(9,6))
ax.plot(x2,y2)
ax.fill_between(x,y,color='g')
ax.fill_between(x2,y2,0,alpha=0.1)
ax.set_xlim([-4,4])
ax.set_title('Campana de gauss')
ax.set_label('Numeros de elementos fuera del promedio')
plt.savefig('gauss.png')
plt.show()
```

#### Importando librerias relevantes

```python
import numpy as np
import pandas as pd
```

#### cargando datasets

```python
marketing_sample1 = pd.read_csv("marketing_campaign_append1.csv")
marketing_sample2 = pd.read_csv("marketing_campaign_append2.csv")
```

```python
marketing_sample1.head()
```

#### Subconjunto de variables relevantes

```python
marketing_sample1 = marketing_sample1[['ID', 'Year_Birth', 'Education', 'Marital_Status',
                'Income', 'Kidhome', 'Teenhome', 'Dt_Customer',
                'Recency','NumStorePurchases', 'NumWebVisitsMonth']]
```

```python
marketing_sample1.head()
```

```python
marketing_sample2 = marketing_sample2[['ID', 'Year_Birth', 'Education', 'Marital_Status',
                'Income', 'Kidhome', 'Teenhome', 'Dt_Customer',
                'Recency','NumStorePurchases', 'NumWebVisitsMonth']]
```

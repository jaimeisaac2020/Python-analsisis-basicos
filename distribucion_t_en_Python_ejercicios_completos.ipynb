La distribución t de Student es una distribución de probabilidad utilizada en estadísticas para inferencia sobre la media de una población cuando el tamaño de la muestra es pequeño o cuando la desviación estándar de la población es desconocida. Aquí tienes 10 ejercicios resueltos de la distribución t de Student en Python:

**Ejercicio 1:**

Calcular la probabilidad de que una variable t de Student con 10 grados de libertad sea menor que 2.

```python
import scipy.stats as stats

grados_de_libertad = 10
valor_t = 2

probabilidad = stats.t.cdf(valor_t, df=grados_de_libertad)
print(f"La probabilidad de que t sea menor que 2 es: {probabilidad:.4f}")
```

Sí, puedes calcular la media, varianza y otros estadísticos de la distribución t de Student en Python utilizando la biblioteca SciPy. Aquí tienes un ejemplo de cómo hacerlo:

```python
import scipy.stats as stats

grados_de_libertad = 10  # Cambia el número de grados de libertad según tu necesidad

# Media de la distribución t de Student
media = stats.t.mean(df=grados_de_libertad)

# Varianza de la distribución t de Student
varianza = stats.t.var(df=grados_de_libertad)

# Mediana de la distribución t de Student
mediana = stats.t.median(df=grados_de_libertad)

# Momento de orden 3 (skewness) de la distribución t de Student
skewness = stats.t.stats(df=grados_de_libertad, moments='s')

# Momento de orden 4 (kurtosis) de la distribución t de Student
kurtosis = stats.t.stats(df=grados_de_libertad, moments='k')

print(f"Media: {media}")
print(f"Varianza: {varianza}")
print(f"Mediana: {mediana}")
print(f"Skewness (Momento de orden 3): {skewness}")
print(f"Kurtosis (Momento de orden 4): {kurtosis}")
```

En este ejemplo, hemos utilizado las funciones `stats.t.mean`, `stats.t.var`, `stats.t.median`, `stats.t.stats` con el argumento `moments` para calcular diferentes estadísticos de la distribución t de Student. Puedes cambiar el valor de `grados_de_libertad` según tus necesidades específicas.

Esto te proporcionará la información estadística básica de la distribución t de Student, como la media, varianza, mediana, skewness (momento de orden 3) y kurtosis (momento de orden 4).

````python
import scipy.stats as stats

grados_de_libertad = 10  # Cambia el número de grados de libertad según tu necesidad

# Media de la distribución t de Student
media = stats.t.mean(df=grados_de_libertad)

# Varianza de la distribución t de Student
varianza = stats.t.var(df=grados_de_libertad)

# Mediana de la distribución t de Student
mediana = stats.t.median(df=grados_de_libertad)

# Momento de orden 3 (skewness) de la distribución t de Student
skewness = stats.t.stats(df=grados_de_libertad, moments='s')

# Momento de orden 4 (kurtosis) de la distribución t de Student
kurtosis = stats.t.stats(df=grados_de_libertad, moments='k')

print(f"Media: {media}")
print(f"Varianza: {varianza}")
print(f"Mediana: {mediana}")
print(f"Skewness (Momento de orden 3): {skewness}")
print(f"Kurtosis (Momento de orden 4): {kurtosis}")
````

```python
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot as plt
```

```python
seed(2021)
x=20*randn(1000)+100
y=x+(10*randn(1000)+50)
```

```python
x
```

```python
plt.scatter(x,y)

plt.show()
```

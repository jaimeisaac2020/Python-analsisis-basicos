2. **Calcular la función de densidad de probabilidad (PDF) y la función de distribución acumulativa (CDF):**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución
media = 0
desviacion_estandar = 1

x = np.linspace(-3, 3, 100)
```

````python
# 2. **Calcular la función de densidad de probabilidad (PDF) y la función de distribución acumulativa (CDF):**


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución
media = 0
desviacion_estandar = 1

# Crear un rango de valores
x = np.linspace(-3, 3, 100)

# Calcular la PDF y CDF
pdf = norm.pdf(x, media, desviacion_estandar)
cdf = norm.cdf(x, media, desviacion_estandar)

# Visualizar la PDF y CDF
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'b-', lw=2, label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)')
plt.xlabel('Valores')
plt.ylabel('Probabilidad acumulada')
plt.legend()

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Parámetros de la distribución binomial
n = 100   # Número de ensayos
p = 0.3   # Probabilidad de éxito en cada ensayo

# Crear una distribución binomial
distribucion_binomial = binom(n, p)

# Calcular la media y la desviación estándar de la distribución binomial
media_binomial = n * p
desviacion_estandar_binomial = np.sqrt(n * p * (1 - p))

# Crear una distribución normal aproximada
distribucion_normal_aproximada = norm(media_binomial, desviacion_estandar_binomial)

# Generar valores para el eje x
x = np.arange(0, n + 1)

# Calcular las probabilidades de la distribución binomial
probabilidades_binomial = distribucion_binomial.pmf(x)

# Calcular las probabilidades de la distribución normal aproximada
probabilidades_normal_aproximada = distribucion_normal_aproximada.pdf(x)

# Graficar ambas distribuciones
plt.figure(figsize=(12, 6))
plt.bar(x, probabilidades_binomial, label='Binomial', alpha=0.5)
plt.plot(x, probabilidades_normal_aproximada, 'r', label='Normal Aproximada', linewidth=2)
plt.title('Aproximación de una Binomial a una Normal')
plt.xlabel('Número de Éxitos')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid()
plt.show()
````
